It is difficult to invest in a company if you do not know where and what kind of activities it is doing. 
In the first place, I do not have a valid passport or visa. So it is necessary to visualize it using electronic data.


```python
import folium
import pandas as pd
```

```python
# Import folium MarkerCluster plugin
from folium.plugins import MarkerCluster
# Import folium MousePosition plugin
from folium.plugins import MousePosition
# Import folium DivIcon plugin
from folium.features import DivIcon
```

```python
# Start location is NVIDIA
coordinate = [37.3706254,-121.9671894]
site_map = folium.Map(location=coordinate, zoom_start=15)
```

```python
circle = folium.Circle(coordinate, radius=20, color='#d35400', fill=True).add_child(folium.Popup('Nvidia'))
# Create a blue circle at NVDIA's coordinate with a icon showing its name
marker = folium.map.Marker(
    coordinate,
    # Create an icon as a text label
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 18; color:#d35400;"><b>%s</b></div>' % 'Semiconductors, Nvidia',
        )
    )
site_map.add_child(circle)
site_map.add_child(marker)
```
